## Range Compression

<img src="./Images/formula/range-Compression.PNG" />

```python
import matplotlib.pyplot as plt
import numpy as np

image = plt.imread('./Images/popular_image.jpg')
image = image[:, :, 2]

# Width & height of image
width, height = image.shape

# Range Compression
output = np.zeros((width, height))
c = 100

output = ( c * np.log10(1 + image)).round().tolist()
```

### Outputs

```python
# Store main image in subplot
plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap="gray")
plt.axis("off")

# Store stretched Image in subplot
plt.subplot(1, 2, 2)
plt.imshow(output, cmap="gray")
plt.axis("off")


plt.show()
```
