## Read a image and generate its histogram

```python
from skimage import exposure
import matplotlib.pyplot as plt

image = plt.imread('./Images/chest_xray_image.png')

# Store the image and its histogram into a fig
plt.figure(figsize=(20, 5))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap="gray")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.hist(image.ravel(), bins=256)

plt.show()
```
