## Contrast Stretching

```python
import numpy as np
from skimage import color
import matplotlib.pyplot as plt

# Read a image
image = plt.imread('./Images/arial_image.png')

# Store main image in subplot
plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1)
plt.imshow(image, cmap="gray")
plt.axis("off")


# Create zeros array to store the stretched image
stretchedImage = np.zeros((image.shape[0], image.shape[1]), dtype = 'uint8')

# Loop over the image and apply Min-Max formulae
maxValue = np.max(image)
minValue = np.min(image)

for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        stretchedImage[i,j] = 255 * (image[i,j] - minValue) / (maxValue - minValue)


# Store stretched Image in subplot
plt.subplot(2, 2, 2)
plt.imshow(stretchedImage, cmap="gray")
plt.axis("off")

# Store the histograms in subplot
plt.subplot(2, 2, 3)
plt.hist(image.ravel(), 256, [0, 256], histtype='step')
plt.subplot(2, 2, 4)
plt.hist(stretchedImage.ravel(), 256, [0, 256], histtype='step')

plt.show()

print(minValue,maxValue)
print(np.min(stretchedImage), np.max(stretchedImage))
```
