## Read a image and generate its histogram

```python
from skimage import exposure
import matplotlib.pyplot as plt
import numpy as np

image = plt.imread('./Images/chest_xray_image.png')

# Store the image and its histogram into a fig
plt.figure(figsize=(20, 5))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap="gray")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.hist(image.ravel(), bins=256)

plt.show()
```

```python
histImage = exposure.equalize_hist(image)

# Show the resulting image
plt.figure(figsize=(20, 5))
plt.subplot(1, 2, 1)
plt.imshow(histImage, cmap="gray")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.hist(histImage.ravel(), bins=256)

plt.show()
```

## Manual way

```python
#Histogram Equalization
hist,bins = np.histogram(img.flatten(),256,[0,256])
cdf = hist.cumsum()

# Find minimum
cdf_min=0
for i in range(cdf.size):
    if(cdf[i]!=0):
        cdf_min=i
        break
        
cdf=(cdf-cdf[cdf_min])*255/(cdf[cdf.size-1]-cdf[cdf_min]).round()
img_cnvt=cdf[img]

# Show the resulting image
plt.figure(figsize=(30, 10))
plt.subplot(2, 2, 1)
plt.imshow(img, cmap="gray")
plt.axis("off")
plt.subplot(2, 2, 2)
plt.hist(img.ravel(), bins=256)

plt.subplot(2, 2, 3)
plt.imshow(img_cnvt, cmap="gray")
plt.axis("off")
plt.subplot(2, 2, 4)
plt.hist(img_cnvt.ravel(), bins=256)
plt.show()
```
