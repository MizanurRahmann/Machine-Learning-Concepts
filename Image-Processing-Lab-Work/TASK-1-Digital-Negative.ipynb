## Read a image and create a histogram for it

```python
import numpy as np
import matplotlib.pyplot as plt

# Read a image
image = plt.imread('./Images/popular_image.jpg')

# Store the image and its histogram into a fig
plt.figure(figsize=(20, 5))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.axis("off")

plt.subplot(1, 2, 2)
plt.hist(image[:, :, 0].ravel(), 256, [0, 256], color="r", histtype='step')
plt.hist(image[:, :, 1].ravel(), 256, [0, 256], color="b", histtype='step')
plt.hist(image[:, :, 2].ravel(), 256, [0, 256], color="g", histtype='step')

# Show the fig
plt.show()

```

## Digital negative

```python
copiedImage = image.copy()
height, width, _ = image.shape

for i in range(0, height - 1):
    for j in range(0, width - 1):
        pixelValue = copiedImage[i, j]
        
        pixelValue[0] = 255 - pixelValue[0]
        pixelValue[1] = 255 - pixelValue[1]
        pixelValue[2] = 255 - pixelValue[2]
        
        copiedImage[i, j] = pixelValue


# Store the image and its histogram into a fig
plt.figure(figsize=(20, 5))
plt.subplot(1, 2, 1)
plt.imshow(copiedImage)
plt.axis("off")

plt.subplot(1, 2, 2)
plt.hist(copiedImage[:, :, 0].ravel(), 256, [0, 256], color="r", histtype='step')
plt.hist(copiedImage[:, :, 1].ravel(), 256, [0, 256], color="b", histtype='step')
plt.hist(copiedImage[:, :, 2].ravel(), 256, [0, 256], color="g", histtype='step')

# Show the fig
plt.show()

            
```

## Compare between input and output

```python
plt.figure(figsize=(20, 20))

plt.subplot(2, 2, 1)
plt.imshow(image)
plt.axis("off")

plt.subplot(2, 2, 2)
plt.imshow(copiedImage)
plt.axis("off")

plt.subplot(2, 2, 3)
plt.hist(image[:, :, 0].ravel(), 256, [0, 256], color="r", histtype='step')
plt.hist(image[:, :, 1].ravel(), 256, [0, 256], color="b", histtype='step')
plt.hist(image[:, :, 2].ravel(), 256, [0, 256], color="g", histtype='step')

plt.subplot(2, 2, 4)
plt.hist(copiedImage[:, :, 0].ravel(), 256, [0, 256], color="r", histtype='step')
plt.hist(copiedImage[:, :, 1].ravel(), 256, [0, 256], color="b", histtype='step')
plt.hist(copiedImage[:, :, 2].ravel(), 256, [0, 256], color="g", histtype='step')

plt.show()
```
