```python
import numpy as np
import matplotlib.pyplot as plt

# Read a image
image = plt.imread('./Images/popular_image.jpg')

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.axis("off")

newImage = image.astype('float')
addedValue = 50
newImage -= addedValue
newImage[newImage < 0] = 0
newImage = newImage.astype('uint8')

plt.subplot(1, 2, 2)
plt.imshow(newImage)
plt.axis("off")

plt.show()
```
