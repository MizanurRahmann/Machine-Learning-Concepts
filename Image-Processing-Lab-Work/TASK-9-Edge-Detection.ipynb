### Read  an image

```python
import numpy as np
import matplotlib.pyplot as plt

# Read the image
image = plt.imread('./Images/popular_image.jpg')
image = image[:, :, 0]
```

### Define a edge detector filter function

```python
def filter_img_using_mask(img, mask):
    img = np.float64(img)
    row, col = img.shape
    output = img.copy()

    for i in range(1, row-1):
        for j in range(1, col-1):
            value = 0
            for k in range(-1, 2):
                for l in range(-1, 2):
                    value = value + (img[i+k][j+l]*mask[k+1][l+1])

            output[i][j] = value

    return output
```

### Define two masks and apply them

```python
#Prewitt operators
po_dx = [[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]
po_dy = [[-1, -1, -1], [0, 0, 0], [1, 1, 1]]

# Sobel operators
so_dx = [[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]]
so_dy = [[-1, -2, -1], [0, 0, 0], [1, 2, 1]]

# laplacian
lo = [[1, 1, 1], [1, -8, 1], [1, 1, 1]]

filteredImage1 = filter_img_using_mask(image, po_dx)
filteredImage2 = filter_img_using_mask(image, po_dy)
combine1 = np.sqrt(np.add(np.square(filteredImage1), np.square(filteredImage2)))

filteredImage3 = filter_img_using_mask(image, so_dx)
filteredImage4 = filter_img_using_mask(image, so_dy)
combine2 = np.sqrt(np.add(np.square(filteredImage3), np.square(filteredImage4)))

filteredImage5 = filter_img_using_mask(image, lo)
```

### Now Show the results

```python
plt.figure(figsize=(20, 10))
plt.subplot(2, 4, 1)
plt.imshow(image, cmap="gray")
plt.title("Original", fontsize=12)
plt.axis("off")

plt.subplot(2, 4, 2)
plt.imshow(filteredImage1, cmap="gray")
plt.title("Prewitt-Vertical", fontsize=12)
plt.axis("off")

plt.subplot(2, 4, 3)
plt.imshow(filteredImage2, cmap="gray")
plt.title("Prewitt-Horizontal", fontsize=12)
plt.axis("off")

plt.subplot(2, 4, 4)
plt.imshow(combine1, cmap="gray")
plt.title("Prewitt-Combine", fontsize=12)
plt.axis("off")

plt.subplot(2, 4, 5)
plt.imshow(image, cmap="gray")
plt.title("Original Image", fontsize=12)
plt.axis("off")

plt.subplot(2, 4, 6)
plt.imshow(filteredImage3, cmap="gray")
plt.title("Sobel-Vertical", fontsize=12)
plt.axis("off")

plt.subplot(2, 4, 7)
plt.imshow(filteredImage4, cmap="gray")
plt.title("Sobel-Horizontal", fontsize=12)
plt.axis("off")

plt.subplot(2, 4, 8)
plt.imshow(combine2, cmap="gray")
plt.title("Sobel-Combine", fontsize=12)
plt.axis("off")

plt.show()


plt.figure(figsize=(20, 10))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap="gray")
plt.title("Original", fontsize=12)
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(filteredImage5, cmap="gray")
plt.title("Prewitt-Vertical", fontsize=12)
plt.axis("off")
plt.show()
```
