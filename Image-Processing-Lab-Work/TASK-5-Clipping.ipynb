## Clipping

<img src="./Images/formula/clipping.PNG" />

```python
import matplotlib.pyplot as plt
import numpy as np

image = plt.imread('./Images/popular_image.jpg')
image = image[:, :, 2]

# Width & height of image
width, height = image.shape

# Clipping
clipped = np.zeros((width, height))
a, b = 50, 150
beta = 255 / ( b - a )

for i in range (width):
    for j in range (height):
        if(image[i,j] < a):
            clipped[i, j] = 0
        elif(image[i,j] >= a and image[i,j] < b):
            clipped[i,j] = beta * (image[i,j] - a)
        else:
            clipped[i, j] = beta * (b - a)
    
```

### Outputs

```python
# Store main image in subplot
plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1)
plt.imshow(image, cmap="gray")
plt.axis("off")

# Store stretched Image in subplot
plt.subplot(2, 2, 2)
plt.imshow(clipped, cmap="gray")
plt.axis("off")

# Store the histograms in subplot
plt.subplot(2, 2, 3)
plt.hist(image.ravel(), 256, [0, 256], histtype='step')
plt.subplot(2, 2, 4)
plt.hist(clipped.ravel(), 256, [0, 256], histtype='step')

plt.show()
```
