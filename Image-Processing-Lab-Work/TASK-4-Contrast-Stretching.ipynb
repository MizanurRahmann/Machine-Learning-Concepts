## Contrast Stretching
Contrast stretching (often called normalization) is a simple image enhancement technique that attempts to improve the contrast in an image by `stretching' the range of intensity values it contains to span a desired range of values,

<img src="./Images/formula/contrast-stretching.PNG" />

```python
import matplotlib.pyplot as plt
import numpy as np

image = plt.imread('./Images/popular_image.jpg')
image = image[:, :, 2]

# Width & height of image
width, height = image.shape

#Contrast Stretching
stretched = np.zeros((width, height))
a, b, ya, yb = 50, 150, 30, 200

alpha= ya / a
beta = (yb - ya) / (b - a)
gama=(255 - yb) / (255 - b)

for i in range (width):
    for j in range (height):
        if(image[i,j] < a):
            stretched[i,j] = image[i,j]*alpha
        elif(image[i,j] >= a and image[i,j]<b):
            stretched[i,j]= (beta*(image[i,j]-a)) + ya
        else:
            stretched[i,j]=(gama*(image[i,j]-b)) + yb


# Store main image in subplot
plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1)
plt.imshow(image, cmap="gray")
plt.axis("off")

# Store stretched Image in subplot
plt.subplot(2, 2, 2)
plt.imshow(stretched, cmap="gray")
plt.axis("off")

# Store the histograms in subplot
plt.subplot(2, 2, 3)
plt.hist(image.ravel(), 256, [0, 256], histtype='step')
plt.subplot(2, 2, 4)
plt.hist(stretched.ravel(), 256, [0, 256], histtype='step')

plt.show()
```
