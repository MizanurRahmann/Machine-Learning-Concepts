```python
import numpy as np
import matplotlib.pyplot as plt

# Read the image
image = plt.imread('./Images/popular_image.jpg')

# Obtaining the red values of the image
red = image[:, :, 0]
# Obtaining the green values of the image
green = image[:, :, 1]
# Obtaining the blue values of the image
blue = image[:, :, 2]
```

```python
plt.imshow(red, cmap="gray")
plt.title('Red')
plt.axis('off')
plt.show()
```

```python
plt.imshow(blue, cmap="gray")
plt.title('Blue')
plt.axis('off')
plt.show()
```

```python
plt.imshow(green, cmap="gray")
plt.title('Green')
plt.axis('off')
plt.show()
```

## Make a Histogram

```python
# Take a array
h = np.zeros(256)

for i in range(0, 512):
    for j in range(0, 512):
        h[ red[i][j] ] += 1

plt.plot(h)
plt.show()
```

## Masking
Get a threshold value and mask the image using the threshold value

```python
threshold = 200

# Create a mask
mask = [[0 for i in range(512)] for j in range(512)]
for i in range (0, 512):
    for j in range(0, 512):
        if red[i][j] > threshold:
            mask[i][j] = 1


# Use mask and mask the red-image
masked_image = red.copy()
for i in range (0, 512):
    for j in range(0, 512):
        if mask[i][j] == 0:
            masked_image[i][j] = 0


# Show the masked image
plt.imshow(masked_image, cmap="gray")
plt.title('Masked Image')
plt.axis('off')
plt.show()
```
